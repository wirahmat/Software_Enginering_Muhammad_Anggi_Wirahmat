Soal:

a = input(), if else, untuk bedain dia kelipatan tiga atau bukan

contoh:

3 -> True

9 -> True

8 -> False

10 -> False


```python
a = input()
a = int(a)

if a % 3 == 0:
  print("{} adalah kelipatan dari 3".format(a))
else:
  print("{} bukan kelipatan dari 3".format(a))
```
